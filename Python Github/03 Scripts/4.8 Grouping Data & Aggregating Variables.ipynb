## 1. Aggregating Variables

```python
# Aggregated mean
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']}) 
```

#### a subset, as its name will only supply insights based on a select number of rows and will not give results of the big picture. running the syntax on the full set will give the most accurate results while looking at everything on a grand scale

## 2. Loyalty flag

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max) 
```

```python
ords_prods_merge.head(15) 
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer' 
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer' 
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer' 
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False) 
```

### Customer types' spending habits

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']}) 
```

## 3. Spending Flag by mean

```python
ords_prods_merge['mean_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean) 
```

```python
ords_prods_merge.head(15) 
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_prices'] > 10, 'spending_flag'] = 'Low spender' 
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_prices'] <= 10, 'spending_flag'] = 'High spender' 
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False) 
```

## 4. Order frequency flag by median

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median) 
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer' 
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] <= 20) & (ords_prods_merge['order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer' 
```

```python
ords_prods_merge.head(15) 
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False) 
```
