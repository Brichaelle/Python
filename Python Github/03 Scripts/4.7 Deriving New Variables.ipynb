## 2. Busiest Day

```python
ords_prods_merge['order_dow'].value_counts(dropna = False) 
```

```python
result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy") 
```

```python
result
```

```python
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False) 
```

```python
# Busiest Days
result = []

for value in ords_prods_merge["order_dow"]:
  if value == 0:
    result.append("Busiest day one")
  elif value == 1:
    result.append("Busiest day two") 
  elif value == 3:
    result.append("Slowest day one")
  elif value == 4:
    result.append("Slowest day two")  
  else:
    result.append("Regularly busy") 
```

```python
result
```

```python
ords_prods_merge['busiest_days'] = result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False) 
```

##### Value counts tell which days are the top two busiest days, 1st and 2nd day of the week, while the slowest days are the 3rd and 4th days

## 3. Busiest Hours

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False) 
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] >= 9) & (ords_prods_merge['order_hour_of_day'] < 16), 'busiest_period_of_day_loc'] = 'Most orders'  
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] <= 6) & (ords_prods_merge['order_hour_of_day'] == 23), 'busiest_period_of_day_loc'] = 'Fewest orders' 
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] >= 17) & (ords_prods_merge['order_hour_of_day'] <= 22), 'busiest_period_of_day_loc'] = 'Average orders' 
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_hour_of_day'] == 7) & (ords_prods_merge['order_hour_of_day'] == 8), 'busiest_period_of_day_loc'] = 'Average order'
```

```python
ords_prods_merge['busiest_period_of_day_loc'].value_counts(dropna = False) 
```

##### Since no futher information was given, missing values for busy hours will be found and resolved

```python
#Looking for missing values
ords_prods_merge.isnull().sum() 
```

```python
#Resolving missing values
ords_prods_merge.shape
```

```python
ords_prods_merge_clean = ords_prods_merge[ords_prods_merge['busiest_period_of_day_loc'].isnull() == False] 
```

```python
ords_prods_merge_clean.shape
```

```python
ords_prods_merge_clean['busiest_period_of_day_loc'].value_counts(dropna = False) 
```
