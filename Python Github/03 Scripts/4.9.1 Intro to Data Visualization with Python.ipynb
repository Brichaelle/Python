# 2. Data Consistency Checks

#### Missing Values

```python
df_customers.isnull().sum() 
```

```python
df_nan = df_customers[df_customers['First Name'].isnull() == True] 
```

```python
df_nan 
```

##### Resolving missing values

```python
df_customers.shape 
```

```python
df_customers_clean = df_customers[df_customers['First Name'].isnull() == False] 
```

```python
df_customers_clean.shape 
```

### Duplicates

```python
df_dups = df_customers_clean[df_customers_clean.duplicated()] 
```

```python
df_dups
```

##### There are no duplicates in the data set

### Mixed-data Types

```python
for col in df_customers_clean.columns.tolist():
  weird = (df_customers_clean[[col]].applymap(type) != df_customers_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_clean[weird]) > 0:
    print (col) 
```

##### There are no mixed-data type variables in the data set
